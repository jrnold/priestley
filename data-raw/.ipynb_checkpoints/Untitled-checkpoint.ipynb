```python
# Imports the Google Cloud client library
from google.cloud import language
from google.cloud.language import enums
from google.cloud.language import types

# Instantiates a client
client = language.LanguageServiceClient()

def add_url(x, client):
    if not x['url']:
        # The text to analyze
        document = types.Document(
            content=x['description'],
            type=enums.Document.Type.PLAIN_TEXT)
        entities = client.analyze_entities(document=document).entities
        people = [entity for entity in entities 
                  if entity.type == enums.Entity.Type.PERSON]
        if people:
            person = people[0]
        url = person.metadata.get("wikipedia_url", None)
        print(f"Found url: {url}")
        x['url'] = url
    return x
```

```python
newbios = [add_url(x, client) for x in bios]
with open("new-bios.json", "w") as f:
    json.dump(newbios, f)
```

```python
bio_urls = {x["text"]: x["url"] for x in newbios if x["url"]}
```

```python
with open("names_from_description_of_the_chart.txt", "r") as f:
    data = f.readlines()
```

```python
import re
out = []

for line in data:
    m = re.search(r"<(.*)>", line)
    if m:
        url = m.group(1)
    else:
        url = None
    text = re.sub(r"(\s*\[.*?\])?(\s*<.*>)?\s*$", "", line).strip()
    if text != "" and not url:
        if text in bio_urls:
            url = bio_urls[text]
            line = line.strip() + " <" + url + ">"
    out.append(line.strip())

with open("Biographies.txt", "w") as f:
    f.write("\n".join(out))
        
```
