```python
ed1764 = open('priestley_bios_1764.txt', 'r')
ed1778 = open("priestley_bios_1778.txt", "r")

lines1778 = set(x.strip() for x in ed1778.readlines())
lines1764 = set(x.strip() for x in ed1764.readlines())


bios_both = list(lines1764 & lines1778)
bios_1764 = [x + ' [1764]' for x in (lines1764 - lines1778)]
bios_1778 = [x + ' [1778]' for x in (lines1778 - lines1764)]

all_bios = sorted(bios_both + bios_1764 + bios_1778)

with open("bios.txt", "w") as f:
    f.write('\n'.join(all_bios))
            
ed1764.close()
ed1778.close()
```
