```python
import json

with open("priestley_bios_1764.json", "r") as f:
    priestly = json.load(f)
```

```python
from spotlight import annotate, SpotlightException, candidates
```

```python
q = """PREFIX db: <http://dbpedia.org/resource/>
PREFIX foaf: <http://xmlns.com/foaf/0.1/> 
PREFIX dbo: <http://dbpedia.org/ontology/> 
SELECT DISTINCT ?person 
WHERE
{{      
  ?person a foaf:Person . 
  ?person dbo:birthDate ?birth .   
  ?person dbo:deathDate ?death . 
  FILTER (?death <= "{death}-01-01"^^xsd:date)
  FILTER (?birth >= "{birth}-01-01"^^xsd:date)
}} 
"""
```

```python
categories = {'J': "Jewish prophet high-priest rabbi",
 "H P.": "Philosopher",
 "F": "Christian",
 "D": "Christian",
 "Moh": "Muslim doctor Mohammadan",
 "Met": "metaphysician",
 "Mor": "moralist",
 "Pol": "political writer",
 "P": "poet",
 "Pa": "painter",
 "St": "statuary",
 "Mu": "musician",
 "Pr": "printer",
 "Act": "actor",
 "Eng": "engraver",
 "Ar": "architect",
 "Or": "orator",
 "Bel": "critic writer Sophist",
 "H": "historian",
 "Geo": "geographer",
 "Ch": "chronologer",
 "Ant": "antiquity",
 "L": "lawyer",
 "M": "mathematician naturalist natural philosopher astronomer",
 "Ph": "physician doctor medical",
 "Chy": "chemist scientist",
 "Engineer": "engineer",
 "Cr": "critic author writer",
 "Po": "Pope",
 "Trav": "traveler",
}

annotations = {}
for person in priestly:
    text = person['name']
    if 'category' in person:
        category = person['category']
        try:
            add_text = categories[category]
        except:
            print(f"Category {person['category']} not found.")
    else:
        add_text = "warrior politician military statesman leader king ruler"
    death = person['end_2'] + 20
    birth = person['start_2'] - 20
    text = f"{person['name']} {add_text}"
    try:
        annote = candidates('http://model.dbpedia-spotlight.org/en/annotate',
                            person['name'],
                            confidence=0.01,
                            filters = {'sparql': q.format(death = death, birth = birth )})
        print(f"Found {person['text']}")
    except SpotlightException:
        print(f"No resources found for {person['text']}")
        continue
    annotations[person['text']] = annote
    

    
    
```

```python
 q.format(death = death, birth = birth )
```

```python
import yaml
data = {k: v[0]['URI'] for k, v in annotations.items()}
with open("annotations_1764.yaml", "w") as fp:
    yaml.dump(data, fp, default_flow_style = False)
```
