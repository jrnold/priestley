```python
import json

with open("priestley_bios_1764.json", "r") as f:
    priestly = json.load(f)
```

```python
from spotlight import annotate, SpotlightException
```

```python
q = """PREFIX db: <http://dbpedia.org/resource/>
PREFIX foaf: <http://xmlns.com/foaf/0.1/> 
PREFIX dbo: <http://dbpedia.org/ontology/> 
SELECT ?birth ?death ?person 
WHERE
{{      
  ?person a dbo:Person . 
  ?person dbo:birthDate ?birth .   
  ?person dbo:deathDate ?death . 
  FILTER (?birth >= "-{birth}-01-01"^^xsd:date) 
  FILTER (?death <= "-{death}-01-01"^^xsd:date)
}} 
"""
```

```python
people = {}
for person in priestly[:100]:
    text = person['name']
    birth = person['start_2'] - 10
    death = person['end_2'] + 10
    try:
        annotations = annotate('http://model.dbpedia-spotlight.org/en/annotate',
                              text = text,
                              filters={'sparql': q.format(birth = birth, death = death)})
        people[person[text]] = annotations
    except SpotlightException:
        print(f"Could not find resources for {text}")
        people[text] = None
    
    
```
