```python
import requests

class Spotlight:
    URL = "https://api.dbpedia-spotlight.org/en"
    
    def __init__(self):
        pass
    
    def spot(self, **kwargs):
        return requests.get(self.URL + '/spot', params = kwargs)
    
    def candidates(self, **kwargs):
        return requests.get(self.URL + '/candidates', params = kwargs)
```
