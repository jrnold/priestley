```python
import yaml
import re
from SPARQLWrapper import SPARQLWrapper, JSON

query = """
PREFIX dbo: <http://dbpedia.org/ontology/>
PREFIX foaf: <http://xmlns.com/foaf/0.1/>
PREFIX yago: <http://dbpedia.org/class/yago/>
SELECT ?person ?birthDate ?deathDate 
WHERE {{
    ?person a dbo:Person ; 
        a yago:WikicatBritishMonarchs .
    OPTIONAL {{? person foaf:birthDate ?birthDate }}
    OPTIONAL {{?person foaf:deathDate ?deathDate }}
    FILTER(CONTAINS(STR(?person), "{name}")) .
}}
"""
with open("../Kings-links.yml", "r") as f:
    kings = yaml.load(f)

sparql = SPARQLWrapper("http://dbpedia.org/sparql")

for king in kings:
    name = king["name"]
    name = re.sub("and.*", "", name)
    ordinal = {"1st": "I", "2d": "II", "3d": "III",
               "4th": "IV", "5th": "V", "6th": "VI",
               "7th": "VII", "8th": "VIII", "9th": "IX",
               "10th": "X", "11th": "XI", "12th": "XII"}
    for k, v in ordinal.items():
        name = re.sub(k, v, name)
    name = name.replace(" ", "_")
    sparql.setQuery(query.format(name = name))
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()

print(results)
```
